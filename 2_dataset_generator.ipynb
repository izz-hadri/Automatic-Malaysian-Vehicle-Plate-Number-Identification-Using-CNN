```
import os
import random
import shutil
from imutils import paths
from sklearn.model_selection import train_test_split
labels = []
```

```
# read image paths and labels
images = list(sorted(paths.list_images('/content/drive/MyDrive/Object Detection/no roboflow/data_preparation/augmented_image')))
annotations = [p.split(os.path.sep)[-1].split(".jpg")[0] for p in images]
for a in annotations:
  label = '/content/drive/MyDrive/Object Detection/no roboflow/data_preparation/augmented_image/annotations/' + a + '.txt'
  labels.append(label)
labels = sorted(labels)

print(images)
print(labels)
```

```
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)

print("Train dataset size:      {}".format(len(X_train)))
print("Validation dataset size: {}".format(len(X_val)))
print("Test dataset size:       {}".format(len(X_test)))
```

```
path_ori = '/content/drive/MyDrive/Object Detection/no roboflow/data_preparation/augmented_image/'
path_ori_annotation = '/content/drive/MyDrive/Object Detection/no roboflow/data_preparation/augmented_image/annotations/'
path_train = '/content/drive/MyDrive/Object Detection/no roboflow/data_preparation/dataset/train/'
path_test = '/content/drive/MyDrive/Object Detection/no roboflow/data_preparation/dataset/test/'
path_valid = '/content/drive/MyDrive/Object Detection/no roboflow/data_preparation/dataset/valid/'
```

```
train_filename = [f.split(os.path.sep)[-1].split(".jpg")[0] for f in X_train]
test_filename = [f.split(os.path.sep)[-1].split(".jpg")[0] for f in X_test]
valid_filename = [f.split(os.path.sep)[-1].split(".jpg")[0] for f in X_val]
print(train_filename)
```

```
for train_image in train_filename:
  shutil.copy2(path_ori + train_image + '.jpg', path_train + train_image + '.jpg')
  shutil.copy2(path_ori_annotation + train_image + '.txt', path_train + train_image + '.txt')
  print(train_image + ' done!')

print('-----------------train copy done-----------------')

for test_image in test_filename:
  shutil.copy2(path_ori + test_image + '.jpg', path_test + test_image + '.jpg')
  shutil.copy2(path_ori_annotation + test_image + '.txt', path_test + test_image + '.txt')
  print(test_image + ' done!')

print('-----------------test copy done-----------------')

for valid_image in valid_filename:
  shutil.copy2(path_ori + valid_image + '.jpg', path_valid + valid_image + '.jpg')
  shutil.copy2(path_ori_annotation + valid_image + '.txt', path_valid + valid_image + '.txt')
  print(valid_image + ' done!')

print('-----------------valid copy done-----------------')
```
